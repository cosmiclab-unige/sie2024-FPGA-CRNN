```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import InputLayer, Reshape, Conv2D, Dropout, MaxPooling2D, AveragePooling2D, Flatten, Softmax,GlobalAveragePooling1D,Dense
from tensorflow.keras.optimizers import Adam
from keras.callbacks import EarlyStopping
from keras.models import load_model
import numpy as np
from tensorflow.keras import activations
from sklearn.metrics import accuracy_score
import random
import pickle
import visualkeras
```

```python
input_shape = (300,8)
model = Sequential()
model.add(InputLayer(input_shape=input_shape))
model.add(Reshape((300, 1, 8)))
model.add(Conv2D(16, kernel_size=(2, 1), padding="same", activation=activations.relu))
model.add(Dropout(0.1))
model.add(MaxPooling2D(pool_size=(2, 1), padding="valid"))
model.add(Conv2D(filters=8, kernel_size=(1, 1), padding="valid", activation=activations.relu))
dim = model.output_shape[1]
model.add(AveragePooling2D(pool_size=(dim, 1)))
model.add(Flatten())
model.add(Softmax())
model.add(visualkeras.SpacingDummyLayer(spacing=100))
```
